# **PREFERENCIAS DE ELECCION SEGUN EDICIÓN **

La Maestría en Ciencia de Datos (Universidad Austral) tiene tres ediciones: Buenos Aires, Rosario y Virtual. En 2025 los alumnos se distribuyeron voluntariamente en la materia "Laboratorio de Implementación I" según la siguiente tabla de frecuencias:

**Edición** | **Gerencial** | **Junior** | **Senior**
:---|:---:|:---:|:---:
Buenos Aires | 2 | 11 | 10
Rosario | 7 | 3 | 3
Virtual | 31 | 17 | 11


**Pregunta:** ¿ Existe diferencia significativa entre las preferencias de esas poblaciones de alumnes?

**ELECCIÓN TEST ESTADÍSTICO**
Se seleccionó el test Chi-cuadrado porque ambas variables analizadas (edición y tipo de materia) son categóricas nominales, sin valor numérico ni orden inherente, lo que descarta métodos basados en medias. Además, este test permite analizar la distribución conjunta de múltiples categorías de dos variables, evaluando si las diferencias observadas en la tabla de contingencia son o no significativas.

PAra el análisis se definieron las siguientes hipótesis:

*   **H0 (Hipótesis nula)**: No existe una diferencia significativa en la distribución del tipo de materia entre las distintas ediciones.
*   **H1 (Hipótesis alternativa):** Existe una diferencia significativa en la distribución del tipo de materia entre las ediciones.

**Criterio de decisión:**
Se adopta un nivel de significancia del 5% (α = 0,05), que es el valor estándar en análisis estadístico.


```r
# Crear la tabla de contingencia con los datos
tabla <- matrix(
  c(2, 11, 10,     # Buenos Aires
    7, 3, 3,       # Rosario
    31, 17, 11),   # Virtual
  nrow = 3,
  byrow = TRUE
)

# Asignar nombres a filas y columnas
rownames(tabla) <- c("Buenos Aires", "Rosario", "Virtual")
colnames(tabla) <- c("Gerencial", "Junior", "Senior")

# Mostrar la tabla
tabla

# Ejecutar el test Chi-cuadrado
resultado <- chisq.test(tabla)

# Mostrar resultado completo
resultado

# Mostrar solo el p-valor (por claridad)
resultado$p.value
```

```r
p <- resultado$p.value

if (p < 0.05) {
  mensaje <- paste0(
    "El p-valor obtenido es ", round(p, 5), ", menor al nivel de significancia del 5%.\n",
    "Por lo tanto, se rechaza la hipótesis nula de independencia.\n",
    "La preferencia de materia depende significativamente de la edición cursada.\n"
  )
} else {
  mensaje <- paste0(
    "El p-valor obtenido es ", round(p, 5), ", mayor al nivel de significancia del 5%.\n",
    "Por lo tanto, no se rechaza la hipótesis nula de independencia.\n",
    "La preferencia de materia es independiente de la edición cursada.\n"
  )
}

cat(mensaje)
```
